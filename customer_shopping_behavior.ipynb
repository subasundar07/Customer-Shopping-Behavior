```python
dataset['Review Rating'] = dataset.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
dataset.isnull().sum()
```

```python
dataset.columns = dataset.columns.str.lower()
dataset.columns = dataset.columns.str.replace(' ','_')
dataset = dataset.rename(columns={'purchase_amount_(usd)':'purchase_amount_usd'})
```

```python
dataset.columns
```

```python
#create a column age group
labels = ["young","adult","midle_age","senior"]
dataset['age_group'] = pd.qcut(dataset['age'],q=4, labels = labels)
```

```python
dataset[['age','age_group']].head(7)
```

```python
#freequency mapping

freequency_mapping={
    "Fortnightly" : 14,
    "Weekly" : 7,
    "Monthly" : 30,
    "Bi-Weekly" : 14,
    "Annually" : 365,
    "Every 3 Month" : 90,
    "Quarterly" :30
}
dataset["purchases_frequency_days"] = dataset[ "frequency_of_purchases"].map(freequency_mapping)
    
```
